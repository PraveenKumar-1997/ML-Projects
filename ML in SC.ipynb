```python
#Load the data
dataset = pd.read_csv('norway_new_car_sales_by_make.csv')
```

```python
dataset.head()
```

```python
# Create a column “Period” with both the Year and the Month
dataset['Period'] = dataset['Year'].astype(str) + '-' + dataset['Month'].astype(str)
```

```python
# Weuse the datetime formatting to make sure format is consistent
dataset['Period'] = pd.to_datetime(dataset['Period']).dt.strftime("%Y-%m")
```

```python
#Create a pivot of the data to show the periods on columns and the car makers on rows 
df = pd.pivot_table(data = dataset, index = 'Make', columns = 'Period', values = 'Quantity', aggfunc = 'sum', fill_value = 0)
```
